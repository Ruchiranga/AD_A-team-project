```python
#the following function will return the total number the cast given the list of actors
def nbr_of_actors(arr):
    if (arr==['No cast available']):
        return 0
    else:
        return len(arr)
```

-Scraping the cast of each movie-

```python
#we initialize our dataframe : Movie_cast 
Movie_cast=[]
#we will iterate on the urls using the tconst variable from the movie dataset to access .
#we will parse the html page of each movie to get the main actors.
#the main web page of each movie gives at most 15 cast members.

for title_id in movies.tconst:
    actor_list=[]
    URL='https://www.imdb.com/title/'+str(title_id)+'/?ref_=fn_al_tt_1'
    req=requests.get(URL)
    body=req.text
    soup=BeautifulSoup(body,'html.parser')
    Cast_table=soup.find('table', class_='cast_list')
    #there is sometimes missing information about the cast specially for old movies.
    if (Cast_table== None):
        actor_list.append('No cast available')
    else:   
        Cast_members=Cast_table.find_all('a')
        for actor in Cast_members:
            if (actor.get('href').startswith('/name/')):
                act_id=actor.get('href')[6:15]
                actor_list.append(act_id)
    #dropping duplicates from the list        
    actor_list=list(dict.fromkeys(actor_list))
    Movie_cast.append({'tconst':title_id,'Cast':actor_list,'Nbr_Main_actors':nbr_of_actors(actor_list)})
   
```

```python
Movie_cast=pd.DataFrame(Movie_cast)    
Movie_cast.head(33) 
```

```python
Actors_awards=[]
for actor_id in actors.nconst:
    URL='https://www.imdb.com/name/'+str(actor_id)+'/awards?ref_=nm_awd'
    req=requests.get(URL)
    body=req.text
    soup=BeautifulSoup(body,'html.parser')
    Awards=soup.find_all('tr')
    if (Awards==None):
        Actors_awards.append({'nconst':actor_id,'Year':None,'category':None,'W/N':None,'Description':None,'Movie':None})
    else:
        
        for award in Awards:
    
            if (award.find('td', class_='award_year')==None):
                year='No year available'
            else:
                year=(award.find('td', class_='award_year')).find('a').text
    
            if (award.find('span',class_='award_category')==None):
                category='no category available'
            else:
                category=award.find('span',class_='award_category').text
        
            if (award.find('td',class_="award_outcome")==None):
                w_n='no result available'
            else:
                w_n=award.find('td',class_="award_outcome").find('b').text
        
        
            if (award.find('td',class_='award_description')==None):
                award_desc='no description available available'
            else:
                award_desc=award.find('td',class_='award_description').text
                if (award.find('td',class_='award_description').find('a')==None):
                    movie='No movie available'
                    tconst=''
                else:
                    movie=(award.find('td',class_='award_description')).find('a').text
                    tconst=(award.find('td',class_='award_description')).find('a').get('href')[7:16]
        
            Actors_awards.append({'nconst':actor_id,'Year':year,'category':category,'W/N':w_n,'Describtion':award_desc,'Movie':movie,'tconst':tconst})
        
    
```

```python
Actors_awards=pd.DataFrame(Actors_awards)
Actors_awards.head()
```

```python
def clean_describ(strr):
    sep='\n'
    if (sep in strr):
        return strr.split(sep)[1].split(sep)[0]
    
```

```python
def clean_year(strr):
    sep='\n'
    if(sep in strr):
        return strr.split(sep)[0]
```

```python
Actors_awards['Year']=Actors_awards.apply(lambda x: clean_year(x['Year']), axis=1)
```

```python
Actors_awards['Describtion']=Actors_awards.apply(lambda x: clean_describ(x['Describtion']), axis=1)
```
